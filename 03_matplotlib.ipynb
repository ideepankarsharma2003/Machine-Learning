# Matplotlib

```python
import matplotlib.pyplot as plt
# pyplot is the module used to create our plots.
```

### `plt.plot()` ---> used to create a plot from the data in a Pandas Series or DataFrame
### `plt.savefig( name&extension )` ---> is used to save an image and display


```python
import pandas as pd 

#  matplotlib works really well with pandas
s = pd.Series([18, 42, 9, 32, 81, 64, 3])  # pandas series
s.plot(kind='bar') # bar plot
plt.savefig('plot.png') # save and display the image
```

## Line Plot 

```python
df = pd.read_csv("https://www.sololearn.com/uploads/ca-covid.csv")
df.head()
```

```python
df.drop('state', axis=1, inplace=True) # drops the state column
df['date'] = pd.to_datetime(df['date'], format="%d.%m.%y")
df['month'] = df['date'].dt.month # add month column
df.set_index('date', inplace=True) # set date as index
df.head()
```

```python
df[df['month'] == 12]['cases'].plot() # plots a `Line Chart` ---> number of cases in December
```

```python
# plots a `Line Chart` ---> number of cases in May and also also include the deaths column
df[df['month'] == 5][['cases', 'deaths']].plot()
```

## Bar Plot

```python
df = pd.read_csv("https://www.sololearn.com/uploads/ca-covid.csv")
df.drop('state', axis=1, inplace=True)
df['date'] = pd.to_datetime(df['date'], format="%d.%m.%y")
df['month'] = df['date'].dt.month
df.set_index('date', inplace=True)

(df.groupby('month')['cases'].sum()).plot(kind="bar") # monthly infection cases
```

```python
(df.groupby('month')['deaths'].sum()).plot(kind='bar')  # monthly death cases
```
